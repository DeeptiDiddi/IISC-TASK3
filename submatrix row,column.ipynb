```python
import numpy as np
a = np.loadtxt("mel_spectrogram.csv", delimiter=",")
M, N = a.shape
print("main matrix shape:", a.shape)
submatrix_files = [
    "submatrix_3.csv",
    "submatrix_3_variant_1.csv",
    "submatrix_3_variant_2.csv",
    "submatrix_3_variant_3.csv",
    "submatrix_3_variant_4.csv"
]
def find_submatrix(a, sub):
    m, n = sub.shape
    for i in range(M - m + 1):
        for j in range(N - n + 1):
            if np.array_equal(a[i:i+m, j:j+n], sub):
                return i, i + m - 1, j, j + n - 1
    return None
for file in submatrix_files:
    sub = np.loadtxt(file, delimiter=",")
    print(f"\n {file} | shape = {sub.shape}")
    result = find_submatrix(a, sub)
    if result:
        r_start, r_end, c_start, c_end = result
        print("FOUND ")
```
