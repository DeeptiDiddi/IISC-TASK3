```python
import numpy as np
from scipy.signal import correlate2d
master = np.loadtxt("mel_spectrogram.csv", delimiter=",")
M, N = master.shape
submatrix_files = [
    "submatrix_3.csv",
    "submatrix_3_variant_1.csv",
    "submatrix_3_variant_2.csv",
    "submatrix_3_variant_3.csv",
    "submatrix_3_variant_4.csv"
]
for file in submatrix_files:
    sub = np.loadtxt(file, delimiter=",")
    m, n = sub.shape
    sub_norm = sub / np.linalg.norm(sub)
    master_norm = master / np.linalg.norm(master)
    corr = correlate2d(master_norm, sub_norm, mode="valid")
    i, j = np.unravel_index(np.argmax(corr), corr.shape)
    print(file, "BEST MATCH")
    print(f"Start Row : {i}")
    print(f"End Row   : {i + m - 1}")
    print(f"Start Col : {j}")
    print(f"End Col   : {j + n - 1}")
    print("Similarity score:", corr[i, j])
```
